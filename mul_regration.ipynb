```python
import numpy as np
x_data1 = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
t_data1 = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]

x_data=np.array(x_data1)
t_data=np.array(t_data1)
print(x_data.shape, t_data.shape)
print(x_data)
print(t_data)
```

```python
w=np.random.rand(3,1)
b=np.random.rand(1)
print(w)
print(b)
```

```python
def loss_func(x,t):
  y=np.dot(x, w)+b
  return (np.sum((t-y)**2))/(len(x))
```

```python
def predict(x):
  y=np.dot(x,w)+b
  return y
```

```python
def numerical_derivative(f,x):
    delta_x=1e-4
    grad=np.zeros_like(x)

    it=np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    
    while not it.finished:
        idx=it.multi_index
        tmp_val=x[idx]
        
        x[idx]=float(tmp_val)+delta_x
        fx1=f(x) # f(x+delta_x)

        x[idx]=float(tmp_val)-delta_x
        fx2=f(x) # f(x-delta_x)

        grad[idx]=(fx1-fx2)/(2*delta_x)

        x[idx]=tmp_val
        it.iternext()

    return grad
```

```python
learning_rate=1e-5
f=lambda x: loss_func(x_data, t_data)
print('Initail error value:',loss_func(x_data, t_data), "w:",w, "b:",b)

for step in range(200001):
  w-=learning_rate*numerical_derivative(f, w)
  b-=learning_rate*numerical_derivative(f, b)

  if step % 400==0:
    print("s:",step,'err', loss_func(x_data, t_data), 'w:',w, 'b:',b)
```
